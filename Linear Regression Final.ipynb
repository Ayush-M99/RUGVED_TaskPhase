```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lr =pd.read_csv(r"C:\Users\Ayush Mishra\Downloads\rugvedlin.csv")
```

```python
lr
```

```python
#Check missing values
missing_values = lr.isnull().sum()
print("Missing values by columns:\n", missing_values)

#remove rows
lr.dropna()

#remove duplicate rows
lr.drop_duplicates()

#fill missing data
lr.fillna(method='ffill')

lr.interpolate()
```

```python
lr
```

```python
X = lr[["depth", "carat"]].copy()
y = lr[["price"]].copy()
```

```python
X
```

```python
y
```

```python
X["intercept"] = 1
```

```python
X = X[["intercept", "depth", "carat"]]
```

```python
X
```

```python
X_T = X.T
```

```python
X_T
```

```python
B = np.linalg.inv(X_T @ X) @ X_T @ y
```

```python
B
```

```python
B.index = X.columns
```

```python
B
```

```python
predictions = X @ B
```

```python
predictions
```

```python
SSR = ((y - predictions) ** 2).sum()
```

```python
SSR
```

```python
TSS = ((y - y.mean()) ** 2).sum()
```

```python
TSS
```

```python
R2 = 1 - (SSR / TSS)
```

```python
R2
```

```python
# Visaulize results
fig, ax = plt.subplots()
ax.scatter(lr["carat"], lr["price"], label="Actual Prices", s=10 )
ax.plot(lr["carat"], predictions, label="Predicted Prices", color="red")
ax.set_xlabel("Carat")
ax.set_ylabel("Price")
ax.set_title("Linear Regression Model for Price vs Carat")
ax.legend()
plt.show()
```
