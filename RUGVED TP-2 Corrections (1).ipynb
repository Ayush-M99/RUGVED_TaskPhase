```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\Ayush Mishra\Documents\matches.csv")
df1 = pd.read_csv(r"C:\Users\Ayush Mishra\Downloads\deliveries.csv")

#16
# Create a new column for season
df1['season'] = df1['match_id'].map(df.set_index('id')['season'])

# Group by season and sum the total runs
runs_df = df1.groupby('season')['total_runs'].sum()

# Print the result
print(runs_df.reset_index())


#18
# Group by season and toss_decision
toss_df = df.groupby(['season', 'toss_decision']).size()

# Unstack the toss_decision column
toss_df = toss_df.unstack('toss_decision')

# Plot the stacked bar chart
toss_df.plot(kind='bar', stacked=True, figsize=(10,6))
plt.xlabel('Season')
plt.ylabel('Number of matches')
plt.title('Toss decision across seasons')
plt.legend(loc='upper right')
plt.show()

#19
# Create a new column for season
df1['season'] = df1['match_id'].map(df.set_index('id')['season'])

# Filter by dismissal_kind
df1 = df1[df1['dismissal_kind'].notnull()]

# Group by season and dismissal_kind
dismissal_df = df1.groupby(['season', 'dismissal_kind']).size()

# Unstack the dismissal_kind column
dismissal_df = dismissal_df.unstack('dismissal_kind')

# Plot the line chart
dismissal_df.plot(figsize=(10,6))
plt.xlabel('Season')
plt.ylabel('Number of dismissals')
plt.title('Dismissal kind across seasons')
plt.legend(loc='upper right')
plt.show()


#20
# Group by batsman and sum the runs
runs_df1 = df1.groupby('batsman').sum()

# Sort the runs in descending order
runs_df1 = runs_df1.sort_values('batsman_runs', ascending=False)

# Select the top 10 batsmen
runs_df1 = runs_df1.head(10)

# Plot the bar chart
runs_df1['batsman_runs'].plot(kind='bar', figsize=(10,6))
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.title('Top 10 run scorers in IPL')
plt.show()

#21


#22
# Concatenate the team1 and team2 columns
teams = pd.concat([df['team1'], df['team2']])

# Get the frequency of each team
freq = teams.value_counts()

# Plot the bar chart
freq.plot(kind='bar', figsize=(10,6))
plt.xlabel('Team')
plt.ylabel('Number of matches')
plt.title('Total number of played matches by each team')
plt.show()

#23

#24
# Get the frequency of each team
freq = df['winner'].value_counts()

# Plot the pie chart
freq.plot(kind='pie', figsize=(10,10), autopct='%1.1f%%')
plt.xlabel('Team')
plt.ylabel('Number of wins')
plt.title('Distribution of won matches')
plt.show()

#25


#26
# Group by toss_winner and toss_decision
toss_df = df.groupby(['toss_winner', 'toss_decision']).size()

# Unstack the toss_decision column
toss_df = toss_df.unstack('toss_decision')

# Plot the stacked bar chart
toss_df.plot(kind='bar', stacked=True, figsize=(10,6))
plt.xlabel('Toss winner')
plt.ylabel('Number of matches')
plt.title('Choice of each team after winning the toss')
plt.legend(loc='upper right')
plt.show()




```
